```python
# Connect to MongoDB
mongo_uri = os.getenv('MONGO_DATABASE_URI')
client = MongoClient(mongo_uri)
db = client.get_default_database()

# Query tickets
tickets = list(db.tickets.find({"assignedTo": ObjectId('67fd9b12ca15f6020474fdbb')}))
print(f"Found {len(tickets)} tickets")
```

```python
# Extract data: assignedByFullNameTitle CREATE description to assignedToFullNameTitle at createdAt
rows = []

for ticket in tickets:
    # Find CREATE history
    create_history = next(
        (h for h in ticket.get('histories', []) if h.get('type') == 'CREATE'), 
        None
    )
    
    if create_history:
        rows.append({
            'ticketNumber': ticket.get('ticketNumberDisplay', str(ticket.get('ticketNumber', ''))),
            'assignedByFullNameTitle': create_history.get('assignedByFullNameTitle', ''),
            'description': ticket.get('description', ''),
            'assignedToFullNameTitle': create_history.get('assignedToFullNameTitle', ''),
            'createdAt': create_history.get('createdAt'),
            'status': ticket.get('status', ''),
            'priority': ticket.get('priorityLevel', '')
        })

df = pd.DataFrame(rows)
df = df.sort_values('createdAt', ascending=False).reset_index(drop=True)
print(f"Processed {len(df)} tickets with CREATE history\n")
df.head(10)
```

```python
# Display in requested format
for _, row in df.iterrows():
    created_at = row['createdAt'].strftime('%Y-%m-%d %H:%M:%S') if pd.notna(row['createdAt']) else 'N/A'
    print(f"[{row['ticketNumber']}] {row['assignedByFullNameTitle']} CREATE {row['description']} to {row['assignedToFullNameTitle']} at {created_at}")
    print()
```

```python
# Summary
print("\n=== SUMMARY ===")
print(f"Total tickets: {len(df)}")
print(f"\nStatus breakdown:")
print(df['status'].value_counts())
print(f"\nAssigned by:")
print(df['assignedByFullNameTitle'].value_counts())
```

```python
# Export
df.to_csv('ca_tickets.csv', index=False)
print("Exported to ca_tickets.csv")
```
